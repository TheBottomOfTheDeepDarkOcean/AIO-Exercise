```python
class CustomLinearRegression:
  def __init__(self, X_data, y_target, learning_rate =0.01, num_epochs = 10000):
    self.num_samples = X_data.shape[0]
    self.y_target = y_target
    self.learning_rate = learning_rate
    self.num_epochs = num_epochs
    self.X_data = X_data

# Initial weights
    self.theta = np.random.randn(self.X_data.shape[1], 1)
    self.losses = []

# Gradient Descent
  def compute_loss(self, y_pred, y_target):
    loss = np.mean((y_pred - y_target)**2)
    return loss

  def predict(self, X_data):
    y_pred = X_data.dot(self.theta)
    return y_pred

  def fit(self):
    for epoch in range(self.num_epochs):
      y_pred = self.predict(self.X_data)
      loss = self.compute_loss(y_pred, self.y_target)
      self.losses.append(loss)
      gradient_theta = (1/self.num_samples) * 2 * self.X_data.T.dot(y_pred - self.y_target)
      self.theta = self.theta - self.learning_rate * gradient_theta

      if( epoch % 50) == 0:
        print(f'Epoch: {epoch} - Loss : {loss}')
    return {
        'loss ': sum(self.losses)/len(self.losses)
    , 'weight': self.theta}
```

```python
linear_model = CustomLinearRegression(X_train_processed, y_train)
linear_model.fit()
```

```python
from sklearn.metrics import r2_score
def r2score(y, y_pred):
  rss = np.sum((y_pred - y)**2)
  tss = np.sum((y - np.mean(y))**2)
  return 1 - (rss/tss)
```

```python
def create_polynomial_features(X, degree =2):
  X_mem = []
  for X_sub in X.T:
    X_sub = X_sub.T
    X_new = X_sub
    for d in range(2, degree +1):
      X_new = np.c_[X_sub, np.power(X_sub, d)]
    X_mem.extend(X_new.T)
  return np.c_[X_mem].T
```
